```python
df['streams'] = pd.to_numeric(df['streams'],errors='coerce')
df.info()
```

```python
max_strem = df['streams'].max()
max_strem

df[df['streams'] == max_strem ]
```

```python
spot = df['in_spotify_charts'].sum()
apple = df['in_apple_charts'].sum()
dee = df['in_deezer_charts'].sum()


total = spot + apple + dee

s_per = spot/total*100
a_per = apple/total*100
d_per = dee/total*100


x = [spot,apple,dee]
label = [f'spotify{s_per}',f'apple{a_per}',f'dezer{d_per}']
fig,ax = plt.subplots()
ax.pie(x,labels=label)
plt.show()
```
